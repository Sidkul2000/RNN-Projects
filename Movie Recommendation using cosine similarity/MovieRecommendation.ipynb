```
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```
df1 = pd.read_csv('moviesData.csv')
df1.head(10)
```

```
df = pd.read_csv('movie_dataset.csv')
df.head()
```

```
def title_index(index):
  return df[df.index == index]['title'].values[0]

def index_title(title):
  return df[df.title == title]['index'].values[0]
```

```
features = ['keywords', 'cast', 'director', 'genres']

def combine_features(row):
  try:
    return row['keywords'] + " " + row['cast'] + " " + row['director'] + " " + row['genres']
  except:
    print("error")

for feature in features:
  df[feature] = df[feature].fillna('')

df['combined_features'] = df.apply(combine_features, axis = 1)

df['combined_features'].head(10)
```

```
cv = CountVectorizer()

cm = cv.fit_transform(df['combined_features'])

cs = cosine_similarity(cm)
```

```
movie_name = input("enter your fav movie: ")
movie_index = index_title(movie_name)
similar = list(enumerate(cs[movie_index]))

sorted_movies = sorted(similar, key = lambda x:x[1], reverse = True)[1:]

a = 0
print("Top 10 movies similar to ", movie_name, " are: ")
for i in sorted_movies:
  print(title_index(i[0]))
  a += 1
  if a>10:
    break
  
```
